```r
sphere_change = function (x) 
{
  n <- nrow(x)
  m <- ncol(x)
  phi <- matrix(0, nrow = n , ncol = m-1)
  for (i in 1:n) {
    for (j in 1:(m-1)) {
      if (j==1){
          phi[i,j] = acos(x[i,j])
        }
      else {
        divisor = 1
        for (k in 1:(j-1)){
          divisor = divisor * sin(phi[i,k])
        }
        phi[i,j] = acos(x[i,j]/divisor)
          }
    }
  }
  return(phi)
}
```

## Coordonnees en $\lambda$

```r
#CHANGEMENT DE VARIABLES A PARTIR DE X
lambda_change = function(x){
  n <- nrow(x)
  m <- ncol(x)
  lambda <- matrix(0, nrow = n , ncol = m-1)
  for (i in 1:n) {
    for (j in 1:(m-1)) {
      if (j==1){
          lambda[i,j] = x[i,j]
        }
      else {
        divisor = 1
        for (k in 1:(j-1)){
          divisor = divisor * (1-lambda[i,k])
        }
        lambda[i,j] = x[i,j]/divisor
          }
    }
  }
  return(lambda)    
}
```

```r
#CHANGEMENT DE VARIABLES A PARTIR DE COORDONNEES SPHERIQUES
lambda_change_2 = function (phi){
    lambda = matrix(0, nrow = nrow(phi), ncol = ncol(phi))
    for(i in 1:nrow(phi)){
        for (j in 1:ncol(phi)){
            lambda[i,j] = cos(phi[i,j])**2
        }
    }
    return(lambda)
}
```

```r
nettoyer = function(x){
    x[x < 0.00001] = 0
    x[is.nan(x)] = 0
    return(x)
}
```
